```python
!pip install openpyxl
import pandas as pd
df_raw = pd.read_excel(r"C:\Users\Acer\Downloads\PYTHON\gdp_pop.xlsx", sheet_name="Data")
df_raw
```

```python
df_raw["Series Name"].unique()
```

```python
df_gdp = df_raw[df_raw["Series Name"] == "GDP (current US$)"]
df_gdp
```

```python
df_pop = df_raw[df_raw["Series Name"] == "Population, total"]
df_pop
```

```python
df_gdp_melt = df_gdp.melt(id_vars=["Country Name", "Country Code"], var_name="Year", value_name="GDP")
df_gdp_melt
```

```python
df_pop_melt = df_pop.melt(id_vars=["Country Name", "Country Code"], var_name="Year", value_name="Population")
df_pop_melt
```

```python
df_gdp_melt['Year'] = df_gdp_melt['Year'].str[:4]
df_pop_melt ['Year'] =df_pop_melt['Year'].str[:4]
```

```python
df_gdp_melt = df_gdp_melt[~df_gdp_melt['Year'].str.contains('seri', case=False, na=False)]
```

```python
df_merged = pd.merge(df_gdp_melt, df_pop_melt, on=["Country Name", "Country Code", "Year"])
df_merged
```

```python
df_merged.replace("..", np.nan, inplace=True)
df_merged= df_merged.dropna()
df_merged
```

```python
df_merged['Year'] = df_merged['Year'].astype(int)
df_merged.dtypes
```

```python
df = df.merge(
    df_merged,
    on=['Year', 'Country Name', 'Country Code'],
    how='left'
)
```

```python
df.duplicated ().sum()
```

```python
df.isnull ().sum()
```

```python
df = df.dropna ()
```

```python
df.dtypes
```

```python
df['GDP per capita'] = df['GDP'] / df['Population']
```

```python
df['Total diseases and injuries'] = np.sum(df[['Injuries','Communicable','NonCommunicable']], axis=1)
```

```python
df = df.round(2)
df
```

```python
df_grouped = df.groupby('Year')[[ 'NonCommunicable', 'Communicable','Injuries',]].mean()

df_grouped.plot(kind='bar', stacked=True, colormap='tab20c', figsize=(12, 4))

plt.title('Average mortality rate from injuries and diseases by year')
plt.xlabel('Year')
plt.ylabel('DALYs')
plt.legend(title='Reasons',loc="center left",  bbox_to_anchor=(1, 0.5) )
plt.tight_layout()
plt.show()
```

The stacked bar chart illustrates the average Disability-Adjusted Life Years (DALYs) per year from three major causes: non-communicable diseases, communicable diseases, and injuries, from 2001 to 2019. This chart provides a clear view of how the burden of these health issues has evolved over time, reflecting global epidemiological trends and healthcare system progress.

Overall, the total number of DALYs from these causes shows a declining trend over the years, indicating advancements in healthcare, medicine, and disease prevention efforts. Non-communicable diseases contribute the largest share of DALYs, with the grey segments consistently occupying the majority of each bar throughout the years. These diseases, such as cardiovascular conditions, diabetes, cancer, and other chronic illnesses, are common in both developed and developing countries, often influenced by modern lifestyle factors like physical inactivity, unhealthy diets, and environmental pollution. As a result, DALYs for this group have shown a slight but steady increase over nearly two decades.

Next, DALYs due to communicable diseases—represented in light green—have shown a downward trend over the years. This reflects global efforts to improve sanitation, expand vaccination programs, promote health education, and strengthen primary healthcare systems. However, this group still contributed a significant portion of DALYs in the early years (2001–2005), indicating that some regions or countries had yet to achieve widespread healthcare coverage at the time.

Lastly, DALYs caused by injuries—shown in blue—have remained relatively stable, consistently making up the smallest proportion among the three categories. Though the percentage is low, injuries continue to cause considerable loss of healthy years, particularly among younger populations and those in the working-age group.
